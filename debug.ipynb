```python
import yfinance as yf
import matplotlib.pyplot as plt
import finplot as fplt
import plotly.graph_objects as go
from datetime import datetime


tickerB3 = "PETR4"
tickerHtml = tickerB3 + ".SA"
assetEntity = yf.Ticker(tickerHtml)
dataAsset = assetEntity.history(period="5d", interval="1m")
dataAsset
if 'Date' not in dataAsset.columns:
    dataAsset['Date'] = dataAsset.index

dataAsset = dataAsset.dropna(subset=['Open', 'High', 'Low', 'Close'])
```

```python
import pandas as pd
dataAsset.to_csv('dataDebug.txt', sep='\t', index=False)
```

```python
fig = go.Figure(data=[go.Candlestick(x=dataAsset['Date'],
                            open=dataAsset['Open'],
                            high=dataAsset['High'],
                            low=dataAsset['Low'],
                            close=dataAsset['Close'])])



fig.show()
```
